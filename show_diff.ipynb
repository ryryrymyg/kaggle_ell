```python
import pandas as pd
import ast
```

```python
diff = pd.read_csv('diff.csv')
```

```python
def loc_str_to_list(str_locs):
    results = []
    for str_loc in str_locs:
        results.append(ast.literal_eval(str_loc))
    results = pd.Series(results)
    return results

def make_null_list(locs):
    results = []
    for loc in locs:
        if loc == ['']:
            loc = []
        results.append(loc)
    results = pd.Series(results)
    return results

def make_sub_anno(sub_locs, pn_hists):
    results = []
    for sub_locs2, pn_hist in zip(sub_locs, pn_hists):
        result = []
        for sub_loc in sub_locs2:
            location=sub_loc.split(" ")
            annotation = pn_hist[int(location[0]):int(location[1])]
            result.append(annotation)
        results.append(result)
    results = pd.Series(results)
    return results

def make_len_col(locs):
    results = []
    for loc in locs:
        results.append(len(loc))
    results = pd.Series(results)
    return results
```

```python
# 必要な行の作成
diff["submission_location"] = loc_str_to_list(diff["submission_location"])
diff["train_location"] = loc_str_to_list(diff["train_location"])
diff["submission_location"] = make_null_list(diff["submission_location"])
diff["train_location"] = make_null_list(diff["train_location"])
diff["submission_annotation"] = make_sub_anno(diff["submission_location"], diff['pn_history'])
diff["submission_location_num"] = make_len_col(diff["submission_location"])
diff["train_location_num"] = make_len_col(diff["train_location"])
```

```python
diff = diff.drop(['case_num', 'pn_num', 'feature_num'], axis=1)
```
